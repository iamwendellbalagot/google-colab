```
%tensorflow_version 1.x
!pip install -q gpt-2-simple
import gpt_2_simple as gpt2
from datetime import datetime
from google.colab import files
```

```
gpt2.download_gpt2(model_name="117M")
```

```
file_name = "POEMS TO TRAIN.txt"
gpt2.copy_file_from_gdrive(file_name)
```

```
sess = gpt2.start_tf_sess()

gpt2.finetune(sess,
              dataset=file_name,
              model_name='117M',
              steps=1000,
              restore_from='fresh',
              run_name='v2',
              print_every=10,
              sample_every=200,
              save_every=500
              )
```

```
gpt2.copy_checkpoint_to_gdrive(run_name='v2')
```

```
gpt2.copy_checkpoint_from_gdrive(run_name = 'v2')
```

```
sess = gpt2.start_tf_sess()
gpt2.load_gpt2(sess, run_name='v2')
```

```
def generate_text(text):
  output = gpt2.generate(sess,
                length=500,
                temperature=1.5,
                prefix= text,
                nsamples=1,
                truncate = '<END>',
                batch_size=1,
                include_prefix = True
                )
  return output  
```

```
generate_text("Loving can hurt")
```
